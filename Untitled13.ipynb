```python
# Import necessary libraries
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
from tensorflow.keras.applications.efficientnet import preprocess_input
import matplotlib.pyplot as plt

# Paths and categories
dataset_path = "/content/drive/MyDrive/dataset2"  # Replace with your dataset path
categories = ["yes", "no"]  # Categories for brain tumor detection

# Load and preprocess the dataset
images = []
labels = []

for category in categories:
    label = 1 if category == "yes" else 0
    category_path = os.path.join(dataset_path, category)
    for file in os.listdir(category_path):
        image_path = os.path.join(category_path, file)
        try:
            # Load and preprocess images
            img = cv2.imread(image_path)
            img = cv2.resize(img, (224, 224))  # Resize to match EfficientNet input size
            images.append(img)
            labels.append(label)
        except Exception as e:
            print(f"Error loading image {image_path}: {e}")

# Convert to NumPy arrays and normalize
images = np.array(images)
labels = np.array(labels)

# Preprocess using EfficientNet's preprocessing function
images = preprocess_input(images)

# Split dataset into 50:50 training and test sets
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.5, random_state=42)

# Further split the training set into training and validation sets
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

print(f"Training set: {X_train_split.shape}, {y_train_split.shape}")
print(f"Validation set: {X_val_split.shape}, {y_val_split.shape}")
print(f"Test set: {X_test.shape}, {y_test.shape}")

# Data augmentation for training data
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    brightness_range=(0.8, 1.2),
    fill_mode='nearest'
)
datagen.fit(X_train_split)

# Build the EfficientNet model
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.6)(x)  # Dropout to reduce overfitting
x = Dense(128, activation='relu', kernel_regularizer=l2(0.01))(x)
x = Dropout(0.5)(x)  # Additional dropout layer
predictions = Dense(1, activation='sigmoid')(x)  # Output layer for binary classification

model = Model(inputs=base_model.input, outputs=predictions)

# Freeze base model layers for transfer learning
for layer in base_model.layers:
    layer.trainable = False

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

print(model.summary())

# Callbacks without EarlyStopping
checkpoint = ModelCheckpoint('best_efficientnet_model.keras', monitor='val_loss', save_best_only=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)

# Train the model
history = model.fit(
    datagen.flow(X_train_split, y_train_split, batch_size=32),
    validation_data=(X_val_split, y_val_split),  # Explicit validation data
    epochs=40,  # Train for all epochs
    callbacks=[checkpoint, reduce_lr]  # Exclude early_stopping
)

# Fine-tune the model
for layer in model.layers:
    layer.trainable = True  # Unfreeze all layers

# Recompile the model for fine-tuning
model.compile(optimizer=Adam(learning_rate=1e-5), loss='binary_crossentropy', metrics=['accuracy'])

# Train the fine-tuned model
history_fine_tune = model.fit(
    datagen.flow(X_train_split, y_train_split, batch_size=32),
    validation_data=(X_val_split, y_val_split),
    epochs=20,  # Train for all epochs
    callbacks=[checkpoint, reduce_lr]  # Exclude early_stopping
)

# Evaluate on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=1)
print(f"Final Test Loss: {test_loss:.4f}")
print(f"Final Test Accuracy: {test_accuracy * 100:.2f}%")

# Plot accuracy and loss
plt.plot(history.history['accuracy'] + history_fine_tune.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'] + history_fine_tune.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title('Model Accuracy')
plt.show()

plt.plot(history.history['loss'] + history_fine_tune.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'] + history_fine_tune.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Model Loss')
plt.show()

# Test a sample image
idx = np.random.randint(0, len(X_test))
sample_image = X_test[idx]
true_label = y_test[idx]

# Predict the class
prediction = model.predict(np.expand_dims(sample_image, axis=0))[0][0]
predicted_label = "Yes" if prediction > 0.5 else "No"

# Display the prediction
plt.imshow(sample_image)
plt.title(f"Prediction: {predicted_label} (True: {'Yes' if true_label == 1 else 'No'})")
plt.axis('off')
plt.show()
```
